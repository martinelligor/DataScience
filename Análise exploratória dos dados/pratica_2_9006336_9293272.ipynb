### Momentos originais

```python
def original_moment(k):
    for column in wine.columns:
        mi_k = 0
        for value in wine[column].unique():
            p_xi = (sum(wine[column]==value))/len(wine[column])
            mi_k += (value**k)*p_xi

        print('%5f' % mi_k)
```

```python
original_moment(1)
```

```python
original_moment(2)
```

```python
original_moment(3)
```

```python
original_moment(4)
```

### Momentos originais

```python
def original_moment(escolas, k):
    escolas = escolas[["MATTOTAL", "NUM_SALAS_EXISTENTES", "NUM_FUNCIONARIOS" ]]
    for column in escolas.columns:
        mi_k = 0
        for value in escolas[column].unique():
            p_xi = (sum(escolas[column]==value))/len(escolas[column])
            mi_k += (value**k)*p_xi

        print(column + ": " + '%5f' % mi_k)
```

```python
original_moment(escolas, 1)
```

```python
original_moment(escolas, 2)
```

```python
original_moment(escolas, 3)
```

```python
original_moment(escolas, 4)
```
