```python
# covarianca por classe
columns = wine.columns.tolist()

plt.figure(figsize=(15, 5))
for c_i, c in enumerate(wine.target.unique()):
    ax = plt.subplot(1, 3, c+1)
    plt.title(c)
    cov = wine.loc[wine.target == c, ['alcohol', 'malic_acid', 'color_intensity', 'hue', 'proline']].cov()
    sns.heatmap(cov)
    
plt.tight_layout()
plt.show()
```

```python
# covarianca por classe
columns = wine.columns.tolist()

plt.figure(figsize=(15, 5))
for c_i, c in enumerate(wine.target.unique()):
    ax = plt.subplot(1, 3, c+1)
    plt.title(c)
    cov = wine.loc[wine.target == c, ['alcohol', 'malic_acid', 'color_intensity', 'hue']].cov()
    sns.heatmap(cov)
    
plt.tight_layout()
plt.show()
```

```python
# covarianca por classe
columns = wine.columns.tolist()

plt.figure(figsize=(15, 5))
for c_i, c in enumerate(wine.target.unique()):
    ax = plt.subplot(1, 3, c+1)
    plt.title(c)
    cov = wine.loc[wine.target == c, ['alcohol', 'malic_acid', 'color_intensity']].cov()
    sns.heatmap(cov)
    
plt.tight_layout()
plt.show()
```

>### A correlação é mais informativa do que a variância, pois, se observamos o plot do heatmap de todas as variáveis, verificamos que a correlação tem uma exibição melhor por possuir seus valores normalizados entre -1 e 1, enquanto que a covariância sofre o efeito da escala dos dados, como pode ser observado no primeiro plot, o qual foi altamente influenciado pelo atributo proline.

>### A correlação seria um método mais indicado para uma análise mais precisa e rápida dos dados, a fim de encontrar variáveis redundantes.

>### A covariância e a correlação auxiliam na visualização de dependências entre as variáveis, para os atributos escolhidos, pode-se observar, por exemplo a relação entre os atributos álcool e proline, na classificação de um vinho, por exemplo. E, de fato, se observadas as variáveis nos boxplots apresentados anteriormente, encontramos uma semelhança entre estas.

```python
cov = escolas[["NUM_SALAS_EXISTENTES", "NUM_FUNCIONARIOS", "NUM_COMPUTADORES", "MATEM", "REDE"]]

columns = hist_MATEM.columns.tolist()
x = 1
plt.figure(figsize=(15, 5))
for c in cov.REDE.unique():
    plt.subplot(1,2,x)
    plt.title(c)
    a = cov.loc[cov.REDE == c].cov()
    sns.heatmap(a)
    x += 1

plt.show()
```

```python
corr = escolas[["NUM_SALAS_EXISTENTES", "NUM_FUNCIONARIOS", "NUM_COMPUTADORES", "MATEM", "REDE"]]

columns = hist_MATEM.columns.tolist()
x = 1
plt.figure(figsize=(20, 7))
for c in corr.REDE.unique():
    plt.subplot(1,2,x)
    plt.title(c)
    a = corr.loc[corr.REDE == c].corr()
    sns.heatmap(a)
    x += 1

plt.show()
```

>#### 1 - Neste caso, a correlação é mais informativa, pois a correlação é um valor entre -1 e 1, ficando mais fácil de verificar se os atributos possuem alguma relação linear entre eles e o quão forte ele é. Na covariância, este processo é mais difícil, já que não sabemos o quão alto ou baixo este valor representa, se ele for positivo, sabemos que quando um valor aumenta o outro também aumenta, mas o quanto o outro aumenta não fica muito claro com este valor.

>#### 2 - A correlação é basicamente a covariância normalizada entre -1 e 1. Se queremos ter um valor numérico que representa a relação linear entre 2 atributos, a correlação é o mais indicado.

>#### 3 - Essas medidas ajudam a verificar se os atributos possuem algum tipo de dependência. No caso das hipóteses feitas, elas podem ajudar a ver se em cada rede de ensino a quantidade de salas, computadores e funcionários cresce igualmente de acordo com o número de alunos matriculados. Podemos ver que em uma escola pública temos uma relação mais linear entre o número de matriculados e o número de salas e funcinários do que em uma escola particulas, algo que não era esperado. Uma forma de verificarmos esta linearidade por ser utilizando o gráfico de dispersão.
